# Lazy Predict

In some cases, we want to apply different machine learning models to a dataset and determine which provides the best results. Manually importing them and training them can prove to be a laborious process. In this situation, the LazyPredict library is useful.

This is an open-source Python library that allows the comparison of various models without adjusting hyperparameters. In a few lines, it is possible to train and compare various machine-learning models. It can be used both for classification and regression problems.

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
```

```python
X = data.data
y= data.target
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.5,random_state =123)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```
