## Data Cleaning

### Removing null values

```python
data['travel_with'] = data['travel_with'].fillna('No travel with info')

data.head()
```

```python
data.isnull().sum()
```

```python
data['total_female'] = data['total_female'].fillna(0)

data['total_male'] = data['total_male'].fillna(0)


data.isnull().sum()
```

```python
data['most_impressing'] = data['most_impressing'].fillna('No comments')


data.isnull().sum()
```

### Checking For And Removing Duplicates

```python
data.duplicated()
```

```python
data.duplicated().sum()
```

```python
# How many rows have been checked for duplicates
data.duplicated().count()
```

### Replacing Characters

```python
data['age_group'] = data['age_group'].str.replace('-',' to ')

data.head()
```
