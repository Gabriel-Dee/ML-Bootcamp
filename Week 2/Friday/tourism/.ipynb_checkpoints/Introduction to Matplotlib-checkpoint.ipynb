### plot x against y values

```python
x=np.arange(-10,11)
y=x**2
print(f'x=\n {x}')
print(f'y=\n {y}')
plt.plot(x,y)  #x in the x-axis; y- in the y-axis
plt.show()
```

### Draw the graph of sin(x) and cos(x) in the same graph

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx)
plt.plot(x,sinx)
plt.show()
```

### Using one plot function for multiple graphs draw the graph of sin(x) and cos(x) in the same graph

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx,x,sinx)  # one plot function in one line
plt.show()
```

### Adding labels to axes. Draw the graph of sin(x) and cos(x) and label the axes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx,x,sinx)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('cosine and sine functions')
plt.show()
```

### Controlling the line style and color


```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx,color='blue',linestyle='-',linewidth=3)
plt.plot(x,sinx, color='red',linestyle='--',linewidth=2.0)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Cosine and sine functions')
plt.show()
```

### Another way of Controlling the line style and color. line style and line style are enclosed in quotation mark

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx,'b-.') #blue color, and also Dash-dot line style
plt.plot(x,sinx,'r--') # red color and dashed line
#plt.plot(x,cosx,'b-.', x,sinx,'r--') # alternatively in one line
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Cosine and sine functions')
plt.show()
```

### Changing fontsize and color

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx, x,sinx)
plt.xlabel('x-values',fontsize=14, color='red')
plt.ylabel('y-values',fontsize=14, color='black')
plt.title('Cosine and sine functions',fontsize=16, color='blue')
plt.show()
```

### Adding markers

```python
x=np.arange(-10,11)
y=x**2
plt.plot(x,y,marker='D') # D=diamond marker. Test with other markers as 
                         # specified in the lecture notes
plt.show()
```

### Adding Grids using plt.grid

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx,'-.b', x,sinx,'--r') 
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Cosine and sine functions')
#plt.grid(True)  # grid on both axis
#plt.grid(True, axis='both') # axis can be 'x', 'y', 'both'
#plt.grid(True, axis='x') # x- axis - vertical lines
#plt.grid(True, axis='y') # y- axis - horizontal lines
plt.grid(True, color='black', linestyle='-', linewidth=) # alternatively: color='k'
plt.show()
# uncomment one at a time (line 8-12 and observe the results)
```

### Adding Legend

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)
plt.plot(x,cosx,label='cos(x)') # label will be used for legend
plt.plot(x,sinx,label='sin(x)')  # label will be used for legend
plt.xlabel('x-values',fontsize=14, color='red')
plt.ylabel('y-values',fontsize=14, color='black')
plt.title('Cosine and sine functions',fontsize=16, color='blue')

#add legend
plt.legend()           #uses default value of location
#plt.legend(loc='best') #default
#plt.legend(loc='upper left') # specify location
#plt.legend(loc='upper right') # specify location
#plt.legend(loc='lower center')

#Remove Frame around the legend
#plt.legend(loc='upper left',frameon=False)

#Number of columns - how legend is spread
#plt.legend(loc='upper left',frameon=False,ncol=2)
#plt.legend(loc='upper left',fancybox=True) #round corners around legend
plt.show()
```

### subplots using plt.subplot

### subplots ( 2 rows, 1 column)

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)

# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
#The commas in the subplot command are optional if numrows*numcols<10
#plt.subplot(211) # is also allowed. 2 rows, 1 column, region 1
fig=plt.figure(layout='constrained') # layout='constrained' required for proper 
                                     # tiles and axis labels
plt.subplot(2, 1, 1) # 2 rows, 1 column, region 1
   
# Make the first plot 
plt.plot(x, cosx) 
plt.title('Cosine')  
plt.grid(True) # show grids

# Set the second subplot as active, and make the second plot. 
#plt.subplot(212) # is also allowed. 2 rows, 1 column, region 2
plt.subplot(2, 1, 2) # 2 rows, 1 column, region 2
plt.plot(x, sinx, color='r') 
plt.title('Sine')  
plt.grid(True) # show grids
plt.show()
```

### subplots ( 1 row, 2 columns)

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)

fig=plt.figure(layout='constrained')
plt.subplot(1, 2, 1) # 1 row, 2 columns, region 1
   
# Make the first plot 
plt.plot(x, cosx) 
plt.title('Cosine')  
plt.grid(True)

# Make the second plot 
plt.subplot(1, 2, 2) # 1 row, 2 columns, region 2
plt.plot(x, sinx, color='r') 
plt.title('Sine')  
plt.grid(True)
plt.show()
```

### subplots ( 2 rows, 2 columns)

```python
#pyplot style
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x ** 2)
y4 = np.cos(x ** 2)

fig=plt.figure(layout='constrained')
plt.subplot(2, 2, 1) # 1st row, 1st column
plt.plot(x, y1) 
plt.title('sin(x)')  
plt.grid(True)

plt.subplot(2, 2, 2) # 1st row, 2nd column
plt.plot(x, y2) 
plt.title('cos(x)')  
plt.grid(True)

plt.subplot(2, 2, 3) # 2nd row, 1st column
plt.plot(x, y3) 
plt.title('sin(x**2)')  
plt.grid(True)

plt.subplot(2, 2, 4) # 2nd row, 2nd column
plt.plot(x, y4) 
plt.title('cos(x**2)')  
plt.grid(True)
plt.show()
```

### Figure size;x and y plotting area limits, x and y axes ticks

```python
# Create a figure of size 8x6 inches, 100 dots per inch(dpi default). 
#default is width=6.4 inches, height=4.8 inches
#Limit x and y min and maximum values

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosx=np.cos(x)
sinx=np.sin(x)

plt.figure(figsize=(8, 6),layout='constrained') # width=8 inches, height=6 inches 
# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(x, cosx, color="blue", linewidth=1.0, linestyle="-",label='cosx')

#Combine figsize in plot
#plt.plot(x, cosx, color="blue", linewidth=1.0, linestyle="-",label='cosx',figsize=(8,6))

# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(x, sinx, color="green", linewidth=1.0, linestyle="-",label='sinx')

# Set x limits
#plt.xlim(-3.2, 3.2)
# Set y limits
plt.ylim(-1.05, 1.04)

#can be done in a single line as:
#plt.axis([xmin, xmax, ymin,ymax])

#plt.axis([-3.2, 3.2,-1.05, 1.04])

# Set x ticks
plt.xticks(np.linspace(-3, 3, 9, endpoint=True))
# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
plt.grid(True)  #add grids
#add legend
plt.legend(loc='upper left') # labels during plot are used with legend
plt.show()
```

# Box plot

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
 
 
# Creating dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)
 
fig = plt.figure(figsize =(10, 7))#10 width & 7 height
 
# Creating plot
plt.boxplot(data)
 
# show plot
plt.show()
```

```python
### Creating dataset
np.random.seed(10)
 
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```

# Violine plot

```python

np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data_to_plot)
plt.show()
```

## series plots

### Series Line plots

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
print(f's.head()\n {s.head()}')
s.plot(title='Matplot Demo') # default is line plot
#s.plot.line(title='Matplot Demo')
plt.grid(True)
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

### Series Vertical Bar plot

```python
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
print(data.head())
#data.plot.bar(alpha=0.7)# default color
#data.plot.bar(alpha=0.7, rot=360) 
data.plot.bar(color='b',alpha=0.7,rot=360)  # specify color
plt.show()
```

### Series Horizontal Bar plot

```python
#Same data as above
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
print(data.head())
data.plot.barh(color='b', alpha=0.6)
plt.show()
```

## Dataframe Line plots

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                   columns=['A', 'B', 'C', 'D'],
                   index=np.arange(0, 100, 10))
print(df.head())
df.plot()   # legend at loc='best'
#df.plot(legend=None)    # Exclude legend
#plt.legend(loc='upper left')  # specify location
plt.title('Simple DataFrame plot')
plt.show()
```

### plot one column versus another using the x and y keywords

```python
# np.random.randn(1000, 2): 1000 rows and 2 columns
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = pd.Series(list(range(len(df3))))
df3.sort_index(axis=1, inplace=True)
print(f'df3.head(10)\n {df3.head(10)}')
df3.plot(x='A', y='B')   # plot column A (x-axis), column B (y-axis)
#df3.plot(x='A', y=['B','C'])   # plot column A (x-axis), columns B and C (y-axis)
plt.show()
```

```python
# plot column A (x-axis), columns B and C (y-axis)
df3.plot(x='A', y=['B','C'])   
plt.show()
```

### Drawing a column from a dataframe - create a series

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=['A', 'B', 'C', 'D'])
print(f'\ndf.A.head()\n {df.A.head()}')
df['A'].plot.bar(alpha=0.8,rot=360)
#alternatively use attribute notation
#df.A.plot.bar(alpha=0.8)
plt.show()
```

### Alternative for plotting a single column

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=['A', 'B', 'C', 'D'])
print(f'\ndf.A.head()\n {df.A.head()}')
df.plot.bar(y='A',alpha=0.8, rot=360)  #x-axis=index, use column name
plt.show()
```

### Plotting two columns

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=['A', 'B', 'C', 'D'])
print(f'df.head()\n {df.head()}')
print(f"\ndf[['A','B']].head()\n {df[['A','B']].head()}")
df[['A','B']].plot.bar(alpha=0.8, rot=360)
plt.show()
```

### Alternative - plotting two columns using parameter y

### Stacked bar plot

## Using  scoresclean.csv

```python
df=pd.read_csv('scoresclean.csv')
print(df.head())
```

### Using value_counts on a Sex column

```python
sexCounts=df.Sex.value_counts()
print(sexCounts)
print(f'\ntype of s is {type(sexCounts)}') # data type is series
print(f'\nsexCounts.index:\n {sexCounts.index}')
print(f'\nsexCounts.values:\n {sexCounts.values}')

#plot the series
df.Sex.value_counts().plot(kind='bar',rot=0)
plt.show()
```

### Using value_counts on Grade Column

```python
df.Grade.value_counts().sort_index().plot(kind='bar',rot=0)
print(f'Grades summary for all Degrees\n {df.Grade.value_counts().sort_index()}')
#plt.grid(axis='y')
plt.show()
```

### Using crosstab on Degree and Grade columns

```python

df2 = pd.crosstab(df["Degree"], df["Grade"])
print(df2)
print("\ntype(df2):",type(df2)) # type is dataframe
print(f'\ndf2.index:\n {df2.index}')
print(f'\ndf2.values:\n {df2.values}')
#plot the results
df2.plot.bar(rot=360)
plt.title('Statistical summary of Degree and Grades')
plt.show()
```

```python
#Another way of representing the information

df2 = pd.crosstab(df['Grade'], df['Degree'])
print(df2)
#plot the results
df2.plot.bar(rot=360)
plt.title('Statistical summary of Degree and Grades')
plt.show()
```
