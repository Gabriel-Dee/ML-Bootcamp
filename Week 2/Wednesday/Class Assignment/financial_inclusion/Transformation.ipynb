```python
# Define the condition for splitting the data 
condition = data['bank_account'] == 'Yes'  

# Subset of data where 'column' < 1000000
new_data = data[condition] 

new_data
```

```python
# Define the condition for splitting the data 
condition = data['cellphone_access'] == 'Yes'  

# Subset of data where 'column' < 1000000
new_data = data[condition] 

new_data
```

```python
# Define the condition for age between 18 and 35
condition = (data['age_of_respondent'] >= 18) & (data['age_of_respondent'] <= 35)

# Subset of data where age is between 18 and 35
new_data = data[condition]

new_data
```

```python
# Convert 'column' to discrete categories
data['discrete_age'] = pd.cut(data['age_of_respondent'], bins=3, labels=['low', 'medium', 'high']) 

# select columns 
columns = ['age_of_respondent','discrete_age']

# show data based on selected columns
data[columns]
```

```python
# Convert 'column' to discrete categories
data['discrete_hsize'] = pd.cut(data['household_size'], bins=3, labels=['low', 'medium', 'high']) 

# select columns 
columns = ['household_size','discrete_hsize']

# show data based on selected columns
data[columns]
```
